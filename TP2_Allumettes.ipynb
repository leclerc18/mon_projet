Nous partirons du principe qu’une partie commence avec 19 allumettes.

Chacun leur tour, les joueurs retirent 1, 2 ou 3 allumettes.

Le joueur a perdu lorsqu'à son tour il ne reste plus qu’une allumette sur le plateau.

**Consignes pour notre version**

- Les joueurs choisissent le nombre d’allumettes en relâchant les touches a, z ou e pour respectivement 1, 2 ou 3 allumettes.

- Pour quitter, ce sera un clic sur la fenêtre.
